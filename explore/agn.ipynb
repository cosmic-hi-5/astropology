```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# directories
data_directory = "/home/edgar/astropology/data/agn_lcs/"
```

```python
# load name of files
with open(f"{data_directory}/lcs.list", "r") as file:

    f_names = file.read()
    f_names = f_names.split("\n")
    f_names.remove("")
```

Summary stats

```python
n_mags = []
median_error = []
min_mags, max_mags, min_max_mags = [], [], []

for f_name in f_names:

    print(f_name, end="\r")

    lc = np.loadtxt(f"{data_directory}/lcs/{f_name}")
    n_mags.append(lc.shape[0])

    min_mag = np.nanmin(lc[:, 1])
    min_mags.append(min_mag)

    max_mag = np.nanmax(lc[:, 1])
    max_mags.append(max_mag)

    min_max_mags.append(max_mag-min_mag)

    median_error.append(np.nanmedian(lc[:, 2]))
    
```

```python
n_bins = 50
_ = plt.hist(min_mags)
```

```python
n_bins = 50
_ = plt.hist(max_mags)
```

```python
n_bins = 50
_ = plt.hist(min_max_mags)
```

```python
n_bins = 50
_ = plt.hist(n_mags)
```

```python
n_bins = 50
_ = plt.hist(median_error)
```
